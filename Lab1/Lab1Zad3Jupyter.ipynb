# Kacper Motyka, ćwiczenie z rysowania wykresów (Lab 1, zadanie 3)

```python
import  matplotlib.pyplot as plt
import numpy as np

#Definicja funkcji, której wykres chcemy wyplotować
f = lambda x: x**2 + 5

def draw(x1, x2, f):
    '''Funkcja, która przyjmuje zakres argumentów oraz funkcję
    i rysuje jej wykres w zadanym zakresie'''
    #Stworzenie wektora x dla zadanego zakresu
    x = np.linspace(x1, x2, 100)

    #Rysowanie wykresu funkcji
    plt.plot(x, f(x))

    #Nadanie tytułu wykresu, nazw osi oraz legendy
    plt.title(f'Wykres funkcji x^2 +5 od {x1} do {x2}')
    plt.xlabel('Argument x')
    plt.ylabel('x^2 + 5')
    plt.legend(['x^2 + 5'])
    plt.show()
```

```python
#Narysowanie wykresu funkcji z zakresie -1, 1
draw(-1, 1, f)
```

```python
#Narysowanie wykresu funkcji z zakresie -6, 6
draw(-6, 6, f)
```

```python
#Narysowanie wykresu funkcji z zakresie 0, 5
draw(0, 5, f)
```
